```python
import pandas as pd

diarios = [
    "CLARIN",
    "PAGINA 12",
    "LA NACION",
    "INFOBAE",
    "PERFIL",
    "CRONISTA",
    "TIEMPO ARGENTINO",
    "AMBITO FINANCIERO",
    "EL POPULAR",
    "LID",
    "PO ultima",
]

dfs = {}

for key in diarios:
    df = pd.read_excel("../data/TITULOS_SENTIMENT ANALYSIS.xlsx", sheet_name=key, header=0)
    df.reset_index(inplace=True)
    print(df.columns)
    print(key)
    df.drop(df.index[:2], inplace=True)
    df.columns = map(str.lower, df.columns)
    dfs[key] = df
```

```python
lid_df = dfs["LID"]

lid_df.drop(lid_df.columns[3:], axis=1, inplace=True)
```

```python
po_df = dfs["PO ultima"]

po_df.drop(po_df.columns[0], axis=1,inplace=True)
```

```python
df = pd.concat(dfs.values(), ignore_index=True)

df = df.loc[df["titulo"].notna()]

df.shape
```

```python
import unidecode
# Van sin tilde

palabras_clave = [
    "Alberto", "Macri", "Mauricio", "CFK", "Cristina", "Del CaÃ±o", "Espert",
    "Centurion", "Pichetto", "Del Pla", "PJ", "peronismo", "FIT", "Cambiemos",
    "Juntos por el Cambio", "Lavagna", "Urtubey", "Rosales", "Hotton"
]

important_indices = []

for i, row in df.iterrows():
    try:
        title = unidecode.unidecode(row["titulo"].lower())
        for clave in palabras_clave:
            if clave.lower() in title:
                print(f"{i} -- {clave} ---> {title}")
                important_indices.append(i)
                break
    except AttributeError as e:
        print(e)
```

```python
len(important_indices)
```

```python
df_politico = df.loc[important_indices]
df_politico.shape
```
